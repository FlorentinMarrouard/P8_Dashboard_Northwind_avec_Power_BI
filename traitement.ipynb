il faut que je fasse une table avec le CA par mois / année 

```python
merged_df = orders.merge(order_details, on='orderID', how='inner')
merged_df
```

```python
# Ajoutez une colonne "total par commande" en multipliant les colonnes "unitPrice" et "quantity"
merged_df['total_par_commande'] = merged_df['unitPrice'] * merged_df['quantity']
merged_df
```

```python
#Assurez-vous que la colonne "orderDate" est de type datetime
merged_df['orderDate'] = pd.to_datetime(merged_df['orderDate'])

# Regroupez les données par date et calculez la somme de la colonne "total par commande"
sum_by_date = merged_df.groupby('orderDate')['total_par_commande'].sum().reset_index()
sum_by_date

# Extrait le mois et l'année de la colonne "orderDate"
sum_by_date['year'] = sum_by_date['orderDate'].dt.year
sum_by_date['month'] = sum_by_date['orderDate'].dt.month

# Regroupez les données par mois et année et calculez la somme de la colonne "total_par_commande"
sum_by_month_and_year = sum_by_date.groupby(['year', 'month'])['total_par_commande'].sum().reset_index()
sum_by_month_and_year
# Le DataFrame "sum_by_month_and_year" contient maintenant la somme de la colonne "total par commande"
# par mois et par année.

# Supposons que vous ayez un DataFrame avec les données
# contenant les colonnes 'year', 'month', et 'total_par_commande'.

# Créez un DataFrame trié par année et mois
sum_by_month_and_year = sum_by_month_and_year.sort_values(by=['year', 'month'])

# Calculez la variation en pourcentage du chiffre d'affaires pour chaque mois par rapport au mois précédent
sum_by_month_and_year['Variation en Pourcentage'] = sum_by_month_and_year['total_par_commande'].pct_change() * 100

# Remplacez les valeurs NaN (pour le premier mois de chaque année) par 0
sum_by_month_and_year['Variation en Pourcentage'].fillna(0, inplace=True)

# Affichez le résultat
sum_by_month_and_year
```

```python


# Supposons que vous ayez un DataFrame avec les données
# contenant les colonnes 'year', 'month', 'total_par_commande', et 'Variation en Pourcentage'.

# Créez un DataFrame trié par année et mois
sum_by_month_and_year = sum_by_month_and_year.sort_values(by=['year', 'month'])

# Ajoutez une nouvelle colonne 'CA du Mois Précédent' qui contient le chiffre d'affaires du mois précédent
sum_by_month_and_year['CA du Mois Précédent'] = sum_by_month_and_year.groupby('year')['total_par_commande'].shift()

# Remplacez les valeurs NaN (pour le premier mois de chaque année) par 0
sum_by_month_and_year['CA du Mois Précédent'].fillna(0, inplace=True)

# Affichez le résultat
sum_by_month_and_year
```

```python
# Supposons que vous avez un DataFrame avec les données
# contenant les colonnes 'year', 'month', 'total_par_commande', 'Variation en Pourcentage', et 'CA du Mois Précédent'.

# Ajoutez une nouvelle colonne 'Variation en Pourcentage (Mois en Cours vs. Mois Précédent)'
sum_by_month_and_year['Variation en Pourcentage (Mois en Cours vs. Mois Précédent)'] = sum_by_month_and_year['total_par_commande'].pct_change() * 100

# Remplacez les valeurs NaN (pour le premier mois de chaque année) par 0
sum_by_month_and_year['Variation en Pourcentage (Mois en Cours vs. Mois Précédent)'].fillna(0, inplace=True)

# Affichez le résultat
sum_by_month_and_year
```

```python
# Supposons que vous avez déjà un DataFrame appelé 'merged_df'

# Convertissez la colonne 'orderDate' en format de date
merged_df['orderDate'] = pd.to_datetime(merged_df['orderDate'])

# Créez une colonne pour extraire l'année et le mois de la commande
merged_df['Year'] = merged_df['orderDate'].dt.year
merged_df['Month'] = merged_df['orderDate'].dt.month

# Groupez les ventes par année et mois, puis comptez-les
sales_by_year_month = merged_df.groupby(['Year', 'Month']).size().reset_index(name='SalesCount')

# Utilisez la méthode shift pour décaler les ventes d'un mois en arrière
sales_by_year_month['PreviousMonthSales'] = sales_by_year_month['SalesCount'].shift(1)

# Calculez la variation en pourcentage par rapport au mois précédent
sales_by_year_month['PercentageChange'] = ((sales_by_year_month['SalesCount'] - sales_by_year_month['PreviousMonthSales']) / sales_by_year_month['PreviousMonthSales']) * 100

# Afficher le DataFrame avec la variation en pourcentage
sales_by_year_month
```
